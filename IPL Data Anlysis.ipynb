```python
balls=pd.read_csv('deliveries.csv')
balls.head(10)
```

```python
batting_tot=balls.groupby('batsman').apply(lambda x:np.sum(x['batsman_runs'])).reset_index(name='Runs')
batting_sorted=batting_tot.sort_values(by='Runs',ascending=False)
batting_sorted2=batting_sorted.head(20)
fig=plt.gcf()
fig.set_size_inches(18.5,8)
plt.xticks(rotation=90)
plt.bar(batting_sorted2['batsman'], batting_sorted2['Runs'],color=['#CD202D','#FFFF48','#292734','#15244C','#005DB7','#EF2920',
               '#FFFF48','#FFFF48','#292734','#FFFF48','#ECC5F2','#EF2920',
               '#292734','#15244C','#005DB7','#005DB7','#292734','#15244C',
               '#FFFF48','#CD202D'])
plt.title('MOST RUNS IN THE IPL',size=22)
plt.show()
```

```python
sixes=balls.groupby('batsman').apply(lambda x:np.sum(x['batsman_runs']==6)).reset_index(name='Sixes')
sixes=sixes.sort_values(by='Sixes',ascending=False)
sixes=sixes.head(20)
fig=plt.gcf()
fig.set_size_inches(18.5,8)
plt.xticks(rotation=90)
plt.bar(sixes['batsman'], sixes['Sixes'],color=['#CD202D','#CD202D','#FFFF48','#FFFF48','#005DB7','#CD202D',
               '#005DB7','#FFFF48','#292734','#FFFF48','#ECC5F2','#EF2920',
               '#292734','#15244C','#005DB7','#005DB7','#292734','#15244C',
               '#005DB7','#CD202D'])
plt.title('MOST SIXES IN THE IPL',size=22)
plt.show()
```

```python
wickets=balls[balls['dismissal_kind']!='run out']
wickets=wickets.groupby('bowler').apply(lambda x:x['dismissal_kind'].dropna()).reset_index(name='Wickets')
wickets=pd.DataFrame(wickets['bowler'].value_counts())
wickets['wickets']=wickets.index
print(wickets.head(5))
wickets2=wickets.head(20)
fig=plt.gcf()
fig.set_size_inches(18.5,8)
plt.xticks(rotation=90)
plt.bar(wickets2['wickets'], wickets2['bowler'],color=['#292734','#CD202D','#FFFF48','#FFFF48','#FFFF48','#292734',
               '#005DB7','#292734','#292734','#FFFF48','#ECC5F2','#EF2920',
               '#292734','#15244C','#CD202D','#FFFF48','#292734','#15244C',
               '#005DB7','#CD202D'])
plt.title('MOST WICKETS IN THE IPL',size=22)
plt.show()
```

```python
runs_given=balls.groupby('bowler').apply(lambda x:np.sum(x['total_runs']-x['penalty_runs']-x['legbye_runs']-x['bye_runs'])).reset_index(name='Runs Conceded')
runs_given=runs_given.sort_values(by='Runs Conceded',ascending=False)
runs_given=runs_given.head(20)
runs_given.head(20)
fig=plt.gcf()
fig.set_size_inches(18.5,8)
plt.xticks(rotation=90)
plt.bar(runs_given['bowler'],runs_given['Runs Conceded'], color=['#FFFF48','#FFFF48','#CD202D','#FFFF48','#CD202D','#292734',
               '#005DB7','#292734','#292734','#FFFF48','#ECC5F2','#EF2920',
               '#292734','#FFFF48','#CD202D','#292734','#292734','#15244C',
               '#005DB7','#CD202D'])
plt.title('MOST RUNS CONCEDED IN THE IPL',size=22)
plt.show()
```

```python
fielder=balls.groupby('fielder').apply(lambda x:x).dropna().reset_index()
fielder=pd.DataFrame(fielder['fielder'].value_counts())
fielder2=fielder.head(15)
fig=plt.gcf()
fig.set_size_inches(18.5,8)
plt.xticks(rotation=60)
plt.bar(fielder2.index,fielder2['fielder'], color=['#FFFF48','#292734','#CD202D','#FFFF48','#CD202D','#292734',
               '#005DB7','#292734','#292734','#CD202D','#ECC5F2','#FFFF48',
               '#FFFF48','#CD202D','#CD202D','#292734','#292734','#15244C',
               '#005DB7','#CD202D'])
plt.title('MOST DISMISSALS IN THE IPL',size=22)
plt.show()
```

```python
batting_factor=1.0
bowling_factor=20.0
fielding_factor=10.0
ar=pd.merge(batting_sorted,wickets,left_on='batsman',right_on='wickets',how='right')
ar=pd.merge(ar,fielder,left_on='batsman',right_on=fielder.index,how='left')

ar=ar.groupby(['batsman','Runs','bowler','fielder']).apply(lambda x:(((x['Runs'])*batting_factor)+((x['bowler'])*bowling_factor)+((x['fielder'])*fielding_factor))).reset_index(name='Total')

top_ar=ar.sort_values(by="Total", ascending=False)
year_wise=top_ar.group
top_ar=top_ar.head(20)
print(top_ar)
fig=plt.gcf()
fig.set_size_inches(18.5,8)
plt.plot(top_ar['batsman'],top_ar['Runs']*batting_factor,'g')
plt.plot(top_ar['batsman'],top_ar['bowler']*bowling_factor,'r')
plt.plot(top_ar['batsman'],top_ar['fielder']*fielding_factor,'y')
plt.plot(top_ar['batsman'],top_ar['Total'])
plt.scatter(top_ar['batsman'],top_ar['Runs']*batting_factor)
plt.scatter(top_ar['batsman'],top_ar['bowler']*bowling_factor)
plt.scatter(top_ar['batsman'],top_ar['fielder']*fielding_factor)
plt.scatter(top_ar['batsman'],top_ar['Total'])
plt.title('MOST IMPACTFUL PLAYERS', SIZE=22)

plt.xticks(rotation=60)
plt.legend(['Run Points','Wicket Points','Dismissal Points','Overall Score'])
plt.show()


```
